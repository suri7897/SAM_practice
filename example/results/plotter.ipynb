```python
import pandas as pd
import matplotlib.pyplot as plt
import os

base_dir = "cnn"
files = {
    "SGD": os.path.join(base_dir, "cnn_SGD_result.csv"),
    "Multistep": os.path.join(base_dir, "cnn_MultiSAM_result.csv"),
    "SAM": os.path.join(base_dir, "cnn_SAM_result.csv"),
}
save_dir = "figures/cnn"
os.makedirs(save_dir, exist_ok=True)

colors = {
    "SGD": "tab:blue",
    "Multistep": "tab:orange",
    "SAM": "tab:green"
}

def plotter(metric_key, title, ylabel, save_name):
    plt.figure(figsize=(6, 4))
    for name, path in files.items():
        df = pd.read_csv(path)
        plt.plot(df["epoch"], df[metric_key], label=f"{name} {metric_key}", color=colors[name])
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel(ylabel)
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"{save_name}.png"), dpi=300)
    plt.show()

plotter("train_loss", "Training Loss", "Loss", "train_loss_plot")
plotter("valid_loss", "Validation Loss", "Loss", "valid_loss_plot")
plotter("train_accuracy", "Training Accuracy", "Accuracy (%)", "train_acc_plot")
plotter("valid_accuracy", "Validation Accuracy", "Accuracy (%)", "valid_acc_plot")
```
